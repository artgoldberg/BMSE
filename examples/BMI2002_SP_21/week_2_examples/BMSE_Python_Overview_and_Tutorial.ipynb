## control flow

```python
if True:
    print(True)
```

```python
if 1<2:
    print('foo')
else:
    print('y')
```

```python
# Python doesn't have a case, or switch statement
age = -7
if age<0:
    print('HUH?')
elif age<12:
    print('kid')
elif age < 19:
    print('teen')
elif age < 60:
    print('adult')
else:
    print('senior')
    
```

```python
# range generates integer sequences; see https://docs.python.org/3/library/stdtypes.html#typesseq-range
for i in range(3,9,2):
    print(i / 4)
```

```python
i=0
while(i<100):
    i+=20
    if 70<i:
        continue
        print(i)
    print(i)
```

## Functions

```python
# functions and methods support required and optional arguments; optional arguments are given default values in the signature
def f(v, t=3, u='x'):
    print((1+v) * t)
```

```python
f(4)
```

```python
# optional arguments can be provided with or without the argument name
f(4,6)
f(4,t=6)
```
